# Über R

R ist eine Programmiersprache, die speziell für die Durchführung statistischer Untersuchungen konzipiert wurde. R ist eine reine Skriptsprache (d.h. Programmcode wird direkt ausgeführt und nicht erst in ein ausführbares Programm umgewandelt/kompiliert); neben dem Einsatz in der Statistik kann R auch für weitere Aufgaben wie das Training von KI-Modellen verwendet werden.

Auf den ersten Blick wirkt R vielleicht komplexer als Excel o.ä. Der große Vorteil von R liegt aber darin, dass die Arbeit mit Daten, die Analyse und auch die Erstellung von aussagekräftigen Grafiken viel leichter möglich ist. Führen Sie z.B. einmal den folgendes Programm durch Klick auf das ▶-Symbol vor dem Codeblock aus:

```r
# Daten laden (hier nutzen wir ein Datenset, dass R bereits mitbringt)
data(airquality)

# Falls Sie sich die Daten anschauen möchten, löschen Sie das '#' vor der folgenden Zeile
# print(airquality)

# Statistische Informationen und Plot
summary(airquality)
plot(airquality)
```

Übrigens: Zeilen die mit einem `#` beginnen, sind Kommentare und werden niemals ausgeführt.

# Wie kann Ich R benutzen?

Immer, wenn Sie auf das ▶-Symbol geklickt haben, haben Sie ein Programm in R ausgeführt. Das Notizbuch übernimmt dabei den technischen Teil, indem ihr Programm in einer R-Umgebung auf den Servern von Google ausgeführt wird.

Sie können R aber auch auf Ihrem eigenen Computer benutzen. Hierzu müssen Sie das Installationspaket von der [offiziellen Webseite](https://www.r-project.org/) herunterladen und ausführen. Anschließend können Sie den Code, den Sie gerade in Codeblöcke in diesem Notizbuch ausgeführt haben, auch auf Ihrem Computer ausführen (in einer Shellumgebung).

In diesem Fall ist aber dringend anzuraten, eine vollwertige Entwicklungsumgebung zu nutzen. Diese unterstützt Sie bei der Arbeit mit R, z.B. durch eine einfachere Anzeige von Fehlern oder Vorschläge beim Programmieren. Eine gute, quelloffene und kostenlose Entwicklungsumgebung für R ist [R-Studio](https://posit.co/products/open-source/rstudio/).
