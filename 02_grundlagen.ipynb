# Variablen

Variablen dienen dazu, Werte - wie das Ergebnis einer Rechnung - zu speichern und später wieder abzurufen. Um einen Wert in einer Variable zu sichern, schreiben wir zunächst den Namen der Variable (diesen können wir frei wählen), gefolgt von `<-` und dem Wert, das wir speichern möchten.

```r
myVarA <- 5 + 7
```

Um auf den Wert der Variable zuzugreifen, schreiben wir einfach den Namen der Variable:

```r
myVarA
```

Variablen können überall dort genutzt werden, wo Sie ansonsten den Wert der Variable verwenden würden. So können wir z.B. die Summe aus `myVarA` und `2` berechnen.

```r
myVarA + 2
```

Einer bereits benutzte Variable können wir jederzeit einen neuen Wert zuweisen, indem wir erneut den Namen gefolgt von `<-` verwenden:

```r
myVarA <- 5
myVarB <- 7
myVarA + myVarB
```

# Vektoren

Variablen können in R verschiedene Typen annehmen. Der gängigste Typ ist der des Vektors; die Variablen `myVarA` und `myVarB` sind Vektoren. Praktischerweise können Vektoren nicht nur einen einzelnen Wert speichern, sondern direkt mehrere Werte hintereinander.

Betrachten wir ein Beispiel:

```r
myVector <- 20
myVector[2] <- 22
myVector[3] <- 24
myVector
```

Die Variable `myVector` enthält also die Werte `20`, `22` und `24` in dieser Reihenfolge. Stellen Sie sich die Variable `myVector` wie einen Warteraum vor. An der vordersten Stelle "wartet" die `20`, auf dem Platz dahinter die `22` und dahinter die `24`:

warteraum.jpg

Wir können nun entweder den gesamten "Warteraum", oder auch nur einzelne "Plätze" abrufen:

```r
myVector[2]
```

Was passiert, wenn wir eine Operation auf den ganzen Vektor (auf den ganzen "Warteraum") anwenden?

```r
myVector + 3
```

Wie Sie sehen, wird die Operation (hier die Addition von `3`) auf jedes Element des Vektors einzeln angewendet.

Es ist sehr mühselig, bei Vektoren wie `myVector` jedesmal alle benötigten Werte hintereinander einzugeben. Glücklicherweise stellt R uns dafür eine Abkürzung bereit. Anstatt jedem Wert explizit einen bestimmten Platz zuzuweisen, können wir folgendes Konstrukt benutzen:

```r
myOtherVector <- 1:3
myOtherVector
```

Der Vektor `myOtherVector` enthält also alle natürlichen Zahlen von `1` bis `3` in dieser Reihenfolge. Auf diese Weise können wir unter Verwendung des `:` schnell Vektoren erstellen, die ganze Zahlenbereiche umfassen.

Dies können wir abschließend nutzen, um uns anzuschauen, was passiert, wenn man zwei Vektoren miteinander addiert:

```r
vectorA <- 1:10
vectorB <- 6:15
vectorA + vectorB
```

Wie Sie sehen, werden die einzelnen Plätze genau ihrer Reihenfolge nach addiert, d.h. `1 + 6`, `2 + 7` etc.

# Dataframes

Dataframes sind ein weiterer Typ von Variablen. Während ein Vektor nur _eine_ Reihe von Werten (wie Zahlen) enthalten kann, kann ein Dataframe _mehrere_ Reihen von Werten enthalten. Dataframes sind besonders dann hilfreich, wenn wir tabellarische Daten in R darstellen möchten.

Schauen wir uns hier als Beispiel einen Dataframe an, den R mitbringt. Rufen wir dafür die Variable `mtcars` auf:

```r
mtcars
```

Der Dataframe `mtcars` enthält Angaben über eine Reihe von Autos in tabellarischer Form; die einzelnen Automodelle stehen in Zeilen, zugehörigen Angaben (Treibstoffverbrauch, Anzahl von Cylindern…) in Spalten.

Einzelne Zellen aus dem Dataframe können wir abrufen, indem wir die gewünschte Zeile und Spalte in eckigen Klammern hinter die Variable des Dataframes schreiben. Dabei können wir sowohl die Nummer der Zeile/Spalte verwenden, wie auch deren Namen:

```r
# Die dritte Spalte der Zeile, die mit "Mazda RX4 Wag" beginnt.
mtcars["Mazda RX4 Wag",3]
```

Technisch gesprochen definieren wir in den eckigen Klammern einen Filter, anhand dessen der gewünschte Wert gesucht und zurückgegeben wird. Diese Funktion ahben wir oben bereits genutzt, als wir einzelne "Plätze" eines Vektors abgerufen haben.

Häufig sind wir aber nicht an einzelnen Werten, sondern direkt an den Werten einer ganzen Spalte interessiert. Glücklicherweise können wir auch diese leicht als Vektor abrufen. Dazu nutzen wir ein `$` gefolgt vom Namen der Spalte:

```r
mtcars$disp
```

# Funktionen

Bisher haben wir nur einfache Rechenoperationen (wie Addition oder Multiplikation) durchgeführt; komplexere Operationen können wir mithilfe von Funktionen bewerkstelligen.

Funktionen werden von R bereitgestellt. Wir nutzen sie, indem wir den Namen der Funktion, gefolgt von runden Klammern verwenden:

```r
print("Die Funktion print() kennen Sie bereits aus der ersten Lektion.")
```

Innerhalb der Klammern können wir Werte an die Funktion übergeben; diese Werte nennen wir Argumente. Argumente können entweder (wie oben) manuell angegeben, oder aber in Form von Variablen übergeben werden. Funktionen sind auch nicht auf ein einzelnes Argument beschränkt, häufig können mehrere Argumente, getrennt durch `,` übergeben werden.

Schauen wir uns dies an, indem wir einige weitere hilfreiche Funktionen testen:

## Kombinieren von Elementen: `c()`

Wie können wir eine Variable definieren, die die Zahlen `1`, `30`, `2`, `1.2` in dieser Reihenfolge enthält? Wie schon oben könnten wir allen Werte explizit einen Platz zuweisen, aber glücklicherweise gibt es eine einfachere Möglichkeit:

```r
myVarD <- c(1, 30, 2, 1.2)
myVarD
```

Die Funktion `c()` kombiniert alle übergebenen Argumente in einem Vektor (`c` steht für _combine_). Hier übergeben wir die Zahlen `1`, `30`, `2` und `1.2`; `c()` liefert uns anschließend einen Vektor, der genau diese Werte enthält.

## Erstellung von Zahlensequenzen: `seq()`

Mittels `seq()` können wir leicht Sequenzen von Zahlen erstellen. Anders als bei der Schreibweise mit `:` können wir dabei nicht nur den Start- und Endpunkt der Sequenz angeben, sondern auch bestimmen in welchem Abstand die einzelnen Zahlen zueinander stehen sollen:

```r
seq(1, 5, by=0.25)
```

Anders als bisher verwenden wir mit `by=0.25` ein optionales Argument. Ohne die Angabe von `by` würde der Abstand `1` betragen:

```r
seq(1, 5)
```

Die meisten Funktionen akzeptieren optionale Argumente. Wir geben diese immer an, indem wir den Namen des Arguments (hier: `by`), gefolgt von einem `=` und dem gewünschten Wert angeben.

Bei notwendigen Argumenten, wie `1` und `5` im obigen Beispiel, können wir uns diesen Mehraufwand sparen. Sie werden über ihre Position innerhalb der Klammern eindeutig identifiziert.

## Filtern: `subset()`

Beim Umgang mit Dataframes passiert es häufig, dass wir nur an den Einträgen interessiert sind, die eine bestimmte Bedingung erfüllen. Wie können wir z.B. alle Autos aus dem Dataframe `mtcars` ausgeben, die weniger als 15 Miles per Gallon (MPG) verbrauchen?

Hierfür können wir die Funktion `subset()` nutzen. Neben unserem Dataframe `mtcars` müssen wir auch unsere Bedingung angeben. Dazu können wir erneut auf den Namen der Spalte in unserem Dataframe zurückgreifen (`mpg`):

```r
subset(mtcars, mpg<15)
```

## Transformieren: `transform()`

Manchmal möchten wir einem bestehenden Dataframe weitere Spalten hinzufügen. So enthält `mtcars` zwar den Verbrauch in _Miles per Gallon_, als Europäer:innen können wir mit diesem Wert aber wenig anfangen. Nützlicher wäre der Verbrauch in l/100km.

Um MPG in l/100km umzurechnen, müssen wir den MPG-Wert durch `235.215` teilen. Fügen wir unserem Dataframe eine neue Spalte `verbrauch` hinzu, die den Verbrauch in l/100km für jedes Modell enthält:

```r
transform(mtcars,verbrauch = 235.215/mpg)
```

Wir nutzen `transform()`, indem wir an die erste Stelle den Dataframe schreiben, gefolgt von einer Gleichung, die den Namen der neuen Spalte als Ergebnis angibt:

    verbrauch = 235.215/mpg

Hier errechnet sich der Verbrauch, indem `235.215` durch den Wert für MPG geteilt wird.

## Hilfe: `help()`

Zu guter Letzt: Die Funktion `help()` zeigt Ihnen die integrierte Hilfe zu einer Funktion an. Diese enthält Informationen zur Verwendung Funktion, listet aber auch alle notwendigen und optionalen Argumente auf.

```r
help(transform)
```
